```python
text = 'අපි යන මඟ දැනටමත් ඉන්නේ!'
text ='ඇවරිගේ සලාදය: කිරිබත්ගොඩ සෙලාන් බැංකුව ගිනි'
```

```python
def load_vocab(vocab=None, vocab_size=None):
    """
    Load vocab from disk. The fisrt four items in the vocab should be <PAD>, <UNK>, <S>, </S>
    """
    # print('Load set vocabularies as %s.' % vocab)
    vocab = [' ' if len(line.split()) == 0 else line.split()[0] for line in codecs.open(vocab, 'r', 'utf-8')]
    vocab = vocab[:vocab_size]
    assert len(vocab) == vocab_size
    word2idx = {word: idx for idx, word in enumerate(vocab)}
    idx2word = {idx: word for idx, word in enumerate(vocab)}
    return word2idx, idx2word
```

```python
dict_file = 'D:\FYP\Dilith-OCR\CDistNet\my_data\images-1000\dict_518.txt'
```

```python
word2idx, idx2word = load_vocab(dict_file, vocab_size=519)
```

```python
import re
chars = []
new_text = text
while len(new_text) > 0:
    for key in word2idx.keys():
        pattern = re.escape(key)
        match = re.match(pattern, new_text)
        if match:
            chars.append(key)
            new_text = re.sub(pattern, '', new_text, count=1)
            break
    print(new_text)
print(chars)
```
